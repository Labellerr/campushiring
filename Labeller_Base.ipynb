```python
import base64
from IPython.display import HTML, display

def show_mp4(path, width=640):
    """
    Reads an MP4 from disk and embeds it as an HTML5 video in Colab.

    Args:
        path (str): path to the .mp4 file
        width (int): display width in pixels
    """
    # Read & encode
    with open(path, 'rb') as f:
        mp4_bytes = f.read()
    b64 = base64.b64encode(mp4_bytes).decode()
    # Build HTML
    html = f'''
    <video width="{width}" controls autoplay loop muted playsinline>
      <source src="data:video/mp4;base64,{b64}" type="video/mp4">
      Your browser does not support HTML5 video.
    </video>
    '''
    display(HTML(html))
```

```python
from ultralytics import YOLO

# Use your trained segmentation model
model = YOLO("/content/drive/MyDrive/best.pt")
```

```python
def bytetrack(path, save_output=True):
    results = model.track(
        source=path,
        tracker="bytetrack.yaml",   # Built-in ByteTrack config
        persist=True,               # Keep consistent IDs
        save=save_output,           # Save output video
        project="/content/drive/MyDrive/yolo_results",  # Save to Drive
        name="bytetrack_output"
    )

    # Extract tracking results frame by frame
    all_tracks = []
    for frame_id, result in enumerate(results):
        if result.boxes is not None and result.boxes.id is not None:
            track_ids = result.boxes.id.int().cpu().tolist()
            bboxes = result.boxes.xyxy.cpu().tolist()
            confidences = result.boxes.conf.cpu().tolist()
            class_ids = result.boxes.cls.int().cpu().tolist()

            for track_id, bbox, conf, cls_id in zip(track_ids, bboxes, confidences, class_ids):
                label = "vehicle" if cls_id == 0 else "pedestrian"
                all_tracks.append({
                    "frame": frame_id,
                    "id": track_id,
                    "class": label,
                    "conf": float(conf),
                    "bbox": [float(x) for x in bbox]
                })
    return all_tracks
```

```python
video_path = "/content/drive/MyDrive/sample_video.mp4"  # your renamed video
tracks = bytetrack(video_path)

# Save JSON results
import json
with open("/content/drive/MyDrive/yolo_results/bytetrack_results.json", "w") as f:
    json.dump(tracks, f, indent=4)

print("âœ… Tracking complete!")
print("Video saved in /content/drive/MyDrive/yolo_results/bytetrack_output/")
print("JSON results saved at /content/drive/MyDrive/yolo_results/bytetrack_results.json")
```

Next, I'll create a simple Streamlit app script. You can modify this script to display the results from your object tracking.

```python
%%writefile app.py
import streamlit as st
import json
import os
from IPython.display import Video, display

st.title("Object Tracking Results")

# Assuming your JSON results and video are in the same directory
json_path = "/content/drive/MyDrive/yolo_results/bytetrack_results.json"
video_path = "/content/drive/MyDrive/yolo_results/bytetrack_demo/sample_video.mp4"

if os.path.exists(json_path):
    with open(json_path, 'r') as f:
        tracking_data = json.load(f)

    st.header("Tracking Data (JSON)")
    st.json(tracking_data)
else:
    st.warning(f"JSON results not found at {json_path}")

st.header("Tracked Video")
if os.path.exists(video_path):
    st.video(video_path)
else:
    st.warning(f"Tracked video not found at {video_path}")
```
