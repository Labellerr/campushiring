```python
import zipfile
import os

with zipfile.ZipFile("campushiring.zip", "r") as zip_ref:
    zip_ref.extractall("CAMPUSHIRING")

```

```python
from ultralytics import YOLO


model = YOLO("yolov8n-seg.pt")
```

```python
yaml_content = """
train: /content/CAMPUSHIRING/campushiring/Image-Segmention-Assignment-Sahil/dataset/images/train
val: /content/CAMPUSHIRING/campushiring/Image-Segmention-Assignment-Sahil/dataset/images/val

nc: 7
names: ['truck', 'pedestrian', 'motorbike', 'auto', 'car', 'cycle', 'bus']
"""

with open("/content/CAMPUSHIRING/campushiring/Image-Segmention-Assignment-Sahil/data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created!")
```

```python
model.train(
    data="/content/CAMPUSHIRING/campushiring/Image-Segmention-Assignment-Sahil/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16
)
```

```python
metrics = model.val()
```

```python
print(metrics.box.map)
```

```python
print(metrics.box.map50)
```

```python
print(metrics.box.map75)
```

```python
print(metrics.confusion_matrix)
```
